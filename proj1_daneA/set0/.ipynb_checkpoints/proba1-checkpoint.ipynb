```python
#from skimage import data
#camera = data.img()
from skimage import filters
edges = filters.scharr(img)
```

```python
io.imshow(edges)
```

```python
from matplotlib import cm    #to wszysko chyba do kitu
image=edges
h, theta, d = hough_line(image)

fig, axes = plt.subplots(1, 3, figsize=(15, 6),
                         subplot_kw={'adjustable': 'box-forced'})
ax = axes.ravel()
ax[0].imshow(image, cmap=cm.gray)  #tto chyba jest rysowanie kresek
for _, angle, dist in zip(*hough_line_peaks(h, theta, d)):
    y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist - image.shape[1] * np.cos(angle)) / np.sin(angle)
    ax[0].plot((0, image.shape[1]), (y0, y1), '-r')
    
ax[0].set_xlim((0, image.shape[1]))
ax[0].set_ylim((image.shape[0], 0))
ax[0].set_axis_off()
ax[0].set_title('Detected lines')

plt.tight_layout()
plt.show()
```
